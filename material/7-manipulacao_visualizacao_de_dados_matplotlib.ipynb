# Matplotlib e Manipulação e Visualização de dados
### Biblioteca para a visualização de dados

`import matplotlib.pyplot as plt`

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Gráfico de linha

```python
# gera 100 dados aleatorios 
alturas = np.round(np.random.normal(1.75, 0.20, 100), 2)
# gera 100 dados aleatorios 
pesos = np.round(np.random.normal(65.50, 15, 100), 2)
```

```python
# cria a figura para o plt
fig, (ax1, ax2) = plt.subplots(ncols=2)

# faz o plot
ax1.plot(alturas, 'o-', linewidth=2)
ax1.grid()
ax1.set_title("Alturas")

ax2.plot(pesos, 'o-', linewidth=2)
ax2.grid()
ax2.set_title("Pesos")

fig.set_figwidth(12)

plt.show()
```

## Manipulação de dados

```python
# carrega os dados no "eleicoes.csv"
dados_eleicoes = np.loadtxt("eleicoes.csv", dtype=str, delimiter=",", skiprows=1)
dados_eleicoes[:11]
```

```python
# pega a quantidade de votos por candidato
votos_goku = dados_eleicoes[dados_eleicoes[:, 0] == "goku"]
votos_vader = dados_eleicoes[dados_eleicoes[:, 0] == "vader"]
votos_yoda = dados_eleicoes[dados_eleicoes[:, 0] == "yoda"]
votos_chuck = dados_eleicoes[dados_eleicoes[:, 0] == "chuck_norris"]

# cria uma lista com os votos
votos = [votos_goku.shape[0], votos_chuck.shape[0], votos_yoda.shape[0], votos_vader.shape[0]]

# calcula a quantidade de votos
qtt_votos = dados_eleicoes.shape[0]

# cria um vetor e calcula a porcentagem de votos
percent = np.array([votos[0], votos[1], votos[2], votos[3]])
percent = ((percent/qtt_votos)*100).round(3)
```

```python
# cria a legenda para o grafico
candidatos = "goku\n{}%; vader\n{}%; yoda\n{}%; chuck_norris\n{}%".format(percent[0],
                                                                 percent[1],
                                                                 percent[2],
                                                                 percent[3])
# separa as legendas em uma lista
candidatos = candidatos.split(";")
```

## Gráfico de Torta (pie)

```python
# cria um grafico pie 
fig, ax = plt.subplots()
votos = [votos_goku.shape[0], votos_chuck.shape[0], votos_yoda.shape[0], votos_vader.shape[0]]
votos1 = [4, 7, 8, 9]

# tamanho 
size = 0.3

# cria cores em rgba
cor_rosa = (1*(233/255), 1*(66/255), 1*(128/255))
cor_lara = (1*(238/255), 1*(121/255), 1*(31/255))
cor_roxa = (1*(75/255), 1*(83/255), 1*(166/255))
cor_cinza = (1*(51/255), 1*(51/255), 1*(51/255))

# cria uma lista de cores
cores = [cor_rosa, cor_cinza, cor_roxa, "c"]

# cria o grafico de pie
ax.pie(votos, radius=1, wedgeprops = dict(width = size, edgecolor = "w"),
       labels = candidatos, colors=cores)


ax.set(
    aspect="equal",
    title='Eleições 2018',
)


plt.show()
```

## Histograma

```python
# cria um grafico de histograma
fig, ax = plt.subplots()
ax.bar(candidatos,height=percent,color=cores)
ax.grid(axis="y", linestyle="-.")
```
